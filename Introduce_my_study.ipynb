### 研究紹介の可視化等

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
x=np.linspace(0,100,101)
y=np.zeros(101)
y[10:25]=np.random.exponential(size=15)+10
y[25:75]=np.random.normal(loc=10,scale=1, size=50)
y[75:90]=np.random.binomial(n=3,p=0.5,size=15)+10
```

```python
plt.figure(figsize=(10,3))
plt.plot(x[0:10],y[0:10],c='black')
plt.plot(x[90:],y[90:],c='black')
plt.plot(x[9:26],y[9:26])
plt.plot(x[25:76],y[25:76])
plt.plot(x[75:91],y[75:91])
```

#### データフレームの制作

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=['A','B','C'])
df
```

```python
y=np.zeros(120)
y[1:25]=np.random.exponential(size=24)+10
y[25:50]=np.random.normal(loc=10,scale=1, size=25)
y[50:80]=np.random.binomial(n=3,p=0.5,size=30)+10
df['A']=y
y[1:25]=np.random.exponential(size=24)+5
y[25:50]=np.random.normal(loc=6,scale=5, size=25)
y[50:80]=np.random.binomial(n=3,p=0.5,size=30)+5
df['B']=y
y[1:25]=np.random.exponential(size=24)-10
y[25:50]=np.random.normal(loc=-10,scale=1, size=25)
y[50:80]=np.random.binomial(n=3,p=0.5,size=30)-10
df['C']=y
```

```python
df.plot(subplots=True,figsize=(8,4),rot=0)
```

```python
y=np.zeros(120)
y[1:30]=np.random.exponential(size=29)+10
y[30:60]=np.random.normal(loc=10,scale=1, size=30)
y[60:110]=np.random.binomial(n=3,p=0.5,size=50)+10
df['A']=y
y[1:30]=np.random.exponential(size=29)+5
y[30:60]=np.random.normal(loc=5,scale=3, size=30)
y[60:110]=np.random.binomial(n=3,p=0.5,size=50)+5
df['B']=y
y[1:30]=np.random.exponential(size=29)-10
y[30:60]=np.random.normal(loc=-10,scale=1, size=30)
y[60:110]=np.random.binomial(n=3,p=0.5,size=50)-10
df['C']=y
```

```python
df.plot(subplots=True,figsize=(8,4),rot=0)
```
